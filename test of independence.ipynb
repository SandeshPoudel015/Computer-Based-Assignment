```python
# importing necessary libraries and commands

import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
```

```python
#Reading the Excel file with crash data and storeing on a variable

crashdata=pd.read_excel("Traffic_crashes.xlsx", sheet_name="Sheet1")
```

```python
#showing some data rows to confirm

crashdata.head(10)
```

```python
# Cross Classifying the data on two basis (Severity and Functional CLass)

field1=crashdata['Severity']
field2=crashdata['Functional_Class']
crosst=pd.crosstab(field1, field2)
crosst
```

```python
# the chi square test
# H0: There is no significant relationship between Crash Severity and Functional Class of Road (the variables are independent)
# H1: There is significant relationship between Crash Severity and Functional Class of Road (the variables are dependent)
```

```python
# determining the values and displaying

chi2,pval,dof,ex_f=chi2_contingency(crosst)
print('Chi Square Value=', chi2)
print('p-value=', pval)
print('Degree of freedom=', dof)
print('Expected Frequency:', ex_f)
```
